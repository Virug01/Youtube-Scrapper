```python
import pandas as pd
import time
import chromedriver_binary
from  selenium import webdriver
import requests
from bs4 import BeautifulSoup 
```

```python
browser = webdriver.Chrome('C:\\Users\\pushk\\chromedriver-win64\\chromedriver.exe')

browser.get('https://www.youtube.com/@tseries/videos')
```

```python
soup = BeautifulSoup(browser.page_source , 'html.parser')
soup
```

```python
data = []

for sp in soup.find_all('ytd-rich-item-renderer'):

    title = sp.find('a', id="video-title-link").text
    link  = "https://www.youtube.com/" + sp.find('a', id="video-title-link").get('href')
    views = sp.find_all(class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[0].text
    time  = sp.find_all(class_ = 'inline-metadata-item style-scope ytd-video-meta-block')[1].text
    try :
        thumbnail = sp.find('img').get('src').split('?')[0]
    except: 
        thumbnail =  None
    data.append([title,views,time,thumbnail,link])
```

```python
data = pd.DataFrame(data, columns = ['Title','Views','Time','Thumbnail','Link'] )
```
